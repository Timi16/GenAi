```python
class ImageGenerator:
    def __init__(self):
        # Use device configuration
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        print(f"Using device: {self.device}")

        # Load the Stable Diffusion pipeline
        self.pipeline = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4")
        self.pipeline.to(self.device)

    def generate_image(self, prompt: str):
        # Generate image from prompt
        with torch.no_grad():
            output = self.pipeline(prompt, num_inference_steps=50)
            image = output.images[0]
        return image
```

```python
# Initialize ImageGenerator
image_generator = ImageGenerator()

# Define your prompt
prompt = "A futuristic cityscape at sunset"

# Generate image
generated_image = image_generator.generate_image(prompt)

# Display the generated image
display(generated_image)
```
